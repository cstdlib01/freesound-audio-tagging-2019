# Visualized data gallery

This solution converts audio into 3 dimentional image.

Followings are examples of beutiful (?) visualized sounds, enjoy.

```python
from dlcliche.notebook import *
from dlcliche.image import *
from lib_fat2019 import *

plt_looks_good()

# Folders
conf.DATA = Path('/mnt/dataset/freesound-audio-tagging-2019')
conf.ROOT = Path('/mnt/dataset/fat2019_files')
conf.WORK = Path('/mnt/dataset/work/fat2019')
```

### APD format is the default

Preprocessed audio is once stored as 1s duration APD. The `load_fat2019_X()` function converts from APD to desired format while loading. It also converts duration of data from 1s to any desired length, and does background subtraction as well.

## AAA format

All channels are amplitude.

```python
conf.DATAMODE = 'AAA'
update_conf(conf)
X = load_fat2019_X(conf, conf.MELSP_TRN_CURATED)
y = pd.read_csv(conf.CSV_TRN_CURATED)
```

```python
for i in range(2):
    plt.imshow(X[i])
    plt.title(y.loc[i].labels)
    plt.show()
```

## APD format

Amplitude, phase and delta of amplitude. You can see higher amplitude part as colored read, and bigger change of amplitude in blue part of image. This is most interesting format.

```python
conf.DATAMODE = 'APD'
update_conf(conf)
X = load_fat2019_X(conf, conf.MELSP_TRN_CURATED)
y = pd.read_csv(conf.CSV_TRN_CURATED)
```

```python
for i in range(2):
    plt.imshow(X[i])
    plt.title(y.loc[i].labels)
    plt.show()
```

## APX format

Amplitude, phase and elementwise multiplication of amplitude and phase.

```python
conf.DATAMODE = 'APX'
update_conf(conf)
X = load_fat2019_X(conf, conf.MELSP_TRN_CURATED)
y = pd.read_csv(conf.CSV_TRN_CURATED)
```

```python
for i in range(2):
    plt.imshow(X[i])
    plt.title(y.loc[i].labels)
    plt.show()
```

## Background subtraction examples

Background sound noise is done by replacing stationary part of sound to zero. This works fine with AAA format only; phase is not processed. Further consideration is needed for phase in the future...

```python
conf.DATAMODE = 'AAA'
conf.DATA_BG = 'SUB_MEAN'

update_conf(conf)
X = load_fat2019_X(conf, conf.MELSP_TRN_CURATED)

for i in range(2):
    plt.imshow(X[i])
    plt.title(y.loc[i].labels)
    plt.show()
```

# Duration conversion examples

```python
conf.DATAMODE = 'APD'
conf.DURATION = 2
conf.DATA_BG = ''
update_conf(conf)
X = load_fat2019_X(conf, conf.MELSP_TRN_CURATED)
y = pd.read_csv(conf.CSV_TRN_CURATED)
plt.imshow(X[0])
plt.title(y.loc[0].labels)
plt.show()

conf.DURATION = 5
update_conf(conf)
X = load_fat2019_X(conf, conf.MELSP_TRN_CURATED)
plt.imshow(X[0])
plt.title(y.loc[0].labels)
plt.show()

conf.DURATION = 10
update_conf(conf)
X = load_fat2019_X(conf, conf.MELSP_TRN_CURATED)
plt.imshow(X[0])
plt.title(y.loc[0].labels);
```
